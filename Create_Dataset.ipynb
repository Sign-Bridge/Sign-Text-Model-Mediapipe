```python
import os
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
```

```python
hands = mp_hands.Hands(static_image_mode=True, min_detection_confidence=0.3)
```

```python
DATA_DIR = 'E:\EAC\Project\Dataset\data' #dtaset directory
```

```python
for directory in os.listdir(DATA_DIR):
    for img_path in os.listdir(os.path.join(DATA_DIR, directory))[:1]:
        img = cv2.imread(os.path.join(DATA_DIR, directory, img_path))
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        results = hands.process(img_rgb)
        if results.multi_hand_landmarks:
            for handLms in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(img_rgb, 
                                          handLms, 
                                          mp_hands.HAND_CONNECTIONS, 
                                          mp_drawing_styles.get_default_hand_landmarks_style(), 
                                          mp_drawing_styles.get_default_hand_connections_style())
                
        plt.figure()
        plt.imshow(img_rgb)
        
plt.show()
```
